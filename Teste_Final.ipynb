```python
resultado_df = data[['sexo','fx_etaria']]
display(resultado_df)
```

```python
resultado_df = data[['sexo','fx_etaria']]
display(resultado_df)
```

```python
df_por_fx_etaria = resultado_df.groupby('fx_etaria')['sexo'].count().reset_index()
df_por_fx_etaria
```

```python
janela = plt.figure(figsize=(10,5))
grafico = janela.add_axes([0,0,1,1])
grafico.bar(df_por_fx_etaria['fx_etaria'], df_por_fx_etaria['sexo'])
plt.xticks(rotation=45)
plt.ylabel('IMIGRANTES')
plt.xlabel('FAIXA ETÁRIA')
plt.title('GRÁFICO DE FAIXA ETÁRIA DE IMIGRANTES')
plt.show()
```

```python
resultado_df1 = data[['sexo','escolaridade']]
display(resultado_df)
```

```python
df_por_escolaridade = resultado_df1.groupby('escolaridade')['sexo'].count().reset_index()
df_por_escolaridade
```

```python
janela = plt.figure(figsize=(10,5))
grafico = janela.add_axes([0,0,1,1])
grafico.bar(df_por_escolaridade['escolaridade'], df_por_escolaridade['sexo'], color ='r')
plt.xticks(rotation=45)
plt.ylabel('IMIGRANTES')
plt.xlabel('FAIXA ETÁRIA')
plt.title('GRÁFICO DE FAIXA POR ESCOLARIDADE')
plt.show()
```
